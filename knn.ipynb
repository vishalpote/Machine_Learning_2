```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
data=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
data
```

```python
data['target']=iris.target
```

```python
data.head()
```

```python
data['flower_name']=data.target.apply(lambda x: data.target.name[x])
```

```python
data.head()
```

```python
from matplotlib import pyplot as plt
```

```python
data0=data[data.target==0]
data0.head()
```

```python
data1=data[data.target==1]
data1
```

```python
data2=data[data.target==2]
data2
```

```python
plt.scatter(data0['sepal length (cm)'],data0['sepal width (cm)'],color="red",marker='+')
plt.scatter(data1['sepal length (cm)'],data1['sepal width (cm)'],color='blue',marker='.')
```

```python
plt.scatter(data0['petal length (cm)'],data0['petal width (cm)'],color="red",marker='+')
plt.scatter(data1['petal length (cm)'],data1['petal width (cm)'],color='blue',marker='.')
```

```python
x=data.drop(['target','flower_name'],axis='columns')
x.head()
```

```python
y=data.target
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
# cm=confusion_matrix()
y_pred=knn.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(5,2))
sn.heatmap(cm,annot='True')
plt.xlabel('y_pred')
plt.ylabel('Truth')
```
