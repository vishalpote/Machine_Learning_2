```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
digit=load_digits()
```

```python
x=digit.data
```

```python
x
```

```python
y=digit.target
y
```

```python
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(x_pca[:,0],x_pca[:,1],c=y)
plt.colorbar()
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
