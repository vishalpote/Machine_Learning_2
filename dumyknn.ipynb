```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
iris=load_iris()
import pandas as pd
```

```python
iris.data
```

```python
iris.target
```

```python
data=pd.DataFrame(iris.data,columns=iris.feature_names)
data.head()
```

```python
data['target']=iris.target
```

```python
data.head()
```

```python
data['flower_names']=data.target.apply(lambda x: iris.target_names)
data.head()
```

```python
data0=data[data.target==0]
data1=data[data.target==1]
data2=data[data.target==2]
```

```python
data0.head()
```

```python
data1.head()
```

```python
data2.head()
```

```python
plt.scatter(data0['sepal length (cm)'],data0['sepal width (cm)'],color="red",marker="+")
plt.scatter(data1['sepal length (cm)'],data1['sepal width (cm)'],color="black",marker="+")
```

```python
plt.scatter(data0['petal length (cm)'],data0['petal width (cm)'],color="red",marker="+")
plt.scatter(data1['petal length (cm)'],data1['petal width (cm)'],color="black",marker="+")
```

```python
x=data.drop(['target','flower_names'],axis='columns')
x.head()
```

```python
y=data.target
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred=knn.predict(x_test)
cm=confusion_matrix(y_pred,y_test)
cm
```

```python
import seaborn as sn 
plt.figure(figsize=(7,5))
sn.heatmap(cm,annot=True)
plt.xlabel(y_pred)
plt.ylabel('truth')
```
